```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
train_df = pd.read_csv("ship.csv")

#a.Use .describe(), .info(), .value_counts()
train_df.info()
train_df.describe(include='all')
for col in train_df.select_dtypes(include=['object']).columns:
    print(f"Value counts for {col}:\n", train_df[col].value_counts(), "\n")

#b.Use sns.pairplot(), sns.heatmap() for visualization
sns.pairplot(train_df.dropna(), hue="Survived")
plt.show()
plt.figure(figsize=(10,8))
# Select only numeric columns
numeric_df = train_df.select_dtypes(include=['number'])
# Now plot the correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Heatmap")
plt.show()

#c.Identify relationships and trends
#   1. Sex vs Survival
sns.barplot(x='Sex', y='Survived', data=train_df)
plt.title("Survival Rate by Sex")
plt.show()
#   2. Pclass (Ticket Class) vs Survival
sns.barplot(x='Pclass', y='Survived', data=train_df)
plt.title("Survival Rate by Passenger Class")
plt.show()
#   3. Age vs Survival
sns.histplot(data=train_df, x='Age', hue='Survived', multiple='stack', bins=30)
plt.title("Survival Distribution by Age")
plt.show()
# 4. Fare vs Survival
sns.boxplot(x='Survived', y='Fare', data=train_df)
plt.title("Fare Distribution by Survival")
plt.show()
#  5. Embarked Location vs Survival
sns.barplot(x='Embarked', y='Survived', data=train_df)
plt.title("Survival Rate by Port of Embarkation")
plt.show()

#d.Plot histograms, boxplots, scatterplots
#1. Histograms  Show distributions of numeric variables


train_df[['Age', 'Fare']].hist(bins=30, figsize=(12, 5), color='skyblue')
plt.suptitle("Histogram of Age and Fare", fontsize=16)
plt.show()
#2. Boxplots  Detect outliers and spread
import seaborn as sns

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.boxplot(x='Survived', y='Age', data=train_df)
plt.title("Age Distribution by Survival")

plt.subplot(1, 2, 2)
sns.boxplot(x='Survived', y='Fare', data=train_df)
plt.title("Fare Distribution by Survival")

plt.tight_layout()
plt.show()
#3. Scatterplots
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Age', y='Fare', hue='Survived', data=train_df)
plt.title("Scatterplot of Age vs Fare by Survival")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.legend(title='Survived')
plt.show()








```
